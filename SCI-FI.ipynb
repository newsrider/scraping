```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get('https://www.goodreads.com/list/show/6934.ScienceFictionBooksbyFemale_Authors')
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
#the rank, book title, book URL, full rating text, and full score text.
#one by one 
#such a clean table 
```

```python
books = doc.find_all('tr')

for book in books:
    ranks = book.find(class_='number')
    print(ranks.text)
```

```python
for book in books:
    booktitle = book.find(class_='bookTitle')
    print(booktitle.text)
    
    
```

```python
for book in books:
    bookurl = book.find('a')
    print("https://www.goodreads.com/"+bookurl['href'])
    
```

```python
for book in books:
    author_name = book.find(class_='authorName')
    print(author_name.text)
```

```python
for book in books:
    rating_text = book.find(class_='minirating')
    print(rating_text.text)
```

```python
for book in books:
    score = book.find(class_='smallText uitext')
    print(score.text)
```

```python
data = []

for book in books: 
    datapoint={}
    datapoint["Rank"]=book.find(class_='number').text
    datapoint["Book"]=book.find(class_='bookTitle').text
    datapoint["URL"]= "https://www.goodreads.com/"+bookurl['href']
    datapoint["Author"]=book.find(class_='authorName').text
    datapoint["Rating Text"]=book.find(class_='minirating').text
    datapoint["Score Text"]=book.find(class_='smallText uitext').text
    data.append(datapoint)

df = pd.DataFrame(data)
df.to_csv("sci-fi.csv", index=False)
```
