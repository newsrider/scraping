```python
driver = webdriver.Chrome()
```

```python
driver.get('http://www.hm.com/us/products/sale/home')
```

```python
while True:
    button = driver.find_element_by_class_name("load-more-btn")
    try:
        button.click()
    except:

        break
```

```python
info = driver.find_elements_by_class_name('product-info')

for index in info:
    print(index.text)
```

```python
#First I used the product-info, so I had a problem that I shared with you in slack,
#then Marie suggested a different class, so it is way more precise. 

data = []
for index in info: 
    datapoint={}
    datapoint["Product Name"] = index.find_element_by_class_name('product-title').text
    datapoint["New Price"] = index.find_element_by_class_name('price').text
    datapoint["Old Price"] = index.find_element_by_class_name('old-price').text
    print(datapoint)
    data.append(datapoint)
    

df = pd.DataFrame(data)
df.to_csv("hm-sales.csv", index=False)
```
